```
# Case 1: 
Assume a plave wave (@28GHz) is incident on a surface with dielectric
 constant 4. How the rflection angle will vary with incident angle.

# Case 2:
Assume a plave wave (@28GHz) is incident on a surface at an angle of 
30 degree. How the rflection angle will vary with the dielectric 
consant of the reflecive surface.

# Case 3:
Reflection cofficient with dielectric constants when 
angle of incidence  = 30 degree.

```





````python
import numpy as np
import matplotlib.pyplot as plt

def reflection_angle(theta_i, n1, n2):
    # reflection angle using Snell's law
    theta_i_radian = np.radians(theta_i)
    sin_theta_r = (n1 / n2) * np.sin(theta_i_radian)
    theta_r = np.degrees(np.arcsin(sin_theta_r))
    return theta_r

# Constants
dielectric_constant = 4 # Dielectric constant of the reflector
freq = 28*10**9  # 28 GHz
c = 3*10**8 # Speed of light
wavelength = c / freq  # in meter

# Dielectric properties
n1 = 1  # Refractive index of the vacuum
n2 = np.sqrt(dielectric_constant)  # Refractive index of the dielectric

# Thickness of the dielectric surface
thickness = 2e-3  # 2 mm

# Range of incidence angles (in degrees)
theta_i = np.linspace(0, 90, 25)

# Calculate reflection angles for different incidence angles
theta_r = reflection_angle(theta_i, n1, n2)

# Plot the results
plt.plot(theta_i, theta_r,'*-')
plt.xlabel('Incidence Angle (degrees)')
plt.ylabel('Reflection Angle (degrees)')
plt.title('Angle of Reflection vs Angle of Incidence \n Dielectric constant of the reflecting surface = 4 ')
plt.grid(True)
plt.show()
````

````python
# Dielectric properties
dielectric_min = 1
dielectric_max = 40
n1 = 1  # Refractive index of the incident medium (usually air/vacuum)
n2_range = np.linspace(np.sqrt(dielectric_min), np.sqrt(dielectric_max), 10)  # Range of dielectric constants

# Angle of incident angle
theta_i = 30  # 30 degrees

# Calculate reflection angles for different dielectric constants
theta_r = reflection_angle(theta_i, n1, n2_range)

# Plot the results
plt.plot(n2_range, theta_r,'r*-')
plt.xlabel('Dielectric Constants')
plt.ylabel('Reflection Angle (degrees)')
plt.title('Angle of Reflection vs Dielectric Constant\n When incidence angle = 30 degree')
plt.grid(True)
plt.show()
````

````python
def reflection_coefficient(n1, n2):
    # reflection coefficient using the refractive indices
    reflection_coefficient = (n1 - n2) / (n1 + n2)
    reflection_coefficient = np.abs(reflection_coefficient) ** 2
    return reflection_coefficient

reflection_coefficients = reflection_coefficient(n1, n2_range)


# Plot
fig, ax1 = plt.subplots()

ax1.plot(n2_range, theta_r, 'r*-')
ax1.set_xlabel('Dielectric Constant')
ax1.set_ylabel('Reflection Angle (degrees)')
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(n2_range, reflection_coefficients, 'b*-')
ax2.set_ylabel('Reflection Coefficient')


plt.title('Angle of Reflection and Reflection Coefficient vs Dielectric Constant')
plt.show()
````
